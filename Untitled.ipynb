```python
import datacube
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
from deafrica_tools.plotting import display_map
from datacube.testutils.io import rio_slurp_xarray
```

```python
dc = datacube.Datacube(app='feature_layers')
```

```python
lat, lon =10.19697, 39.3860

buffer = 0.1

lon_range = (lon - buffer, lon + buffer)
lat_range = (lat + buffer, lat - buffer)
```

```python
display_map(lon_range, lat_range)
```

```python
query = {
    'time':'2019',
    'measurements': ['mask'],
    'resolution': (-20, 20),
    'output_crs': 'epsg:6933',
    'resampling':'nearest',
    'x': lon_range,
    'y': lat_range,
}
```

```python
cm = dc.load('crop_mask_eastern', **query)
```

```python
cm.mask.plot()
```

```python
url_slope = "https://deafrica-input-datasets.s3.af-south-1.amazonaws.com/srtm_dem/srtm_africa_slope.tif"
slope = rio_slurp_xarray(url_slope, gbox=cm.geobox)
```

```python
slope.where(cm.mask==0).squeeze().plot.imshow(size=10, vmax=70)
```

```python
(slope < 50).plot(size=10)
```
