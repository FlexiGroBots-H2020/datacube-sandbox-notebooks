## Load and view Landsat data

```python
#Create a query object
query = {
    'x': lon_range,
    'y': lat_range,
    'time': time,
    'measurements': ['red','nir'],
    'resolution': (-30, 30),
    'group_by': 'solar_day',
    'output_crs': 'EPSG:6933'
}
```

```python
# load cloud-masked ls8 using load_ard
ds_ls8 = load_ard(dc=dc,
              products=['ls8_sr'],
              min_gooddata=0.95,
              mask_filters=(['opening',5], ['dilation',5]),
              dask_chunks=dask_chunks,
              **query,
             )
```

```python
# load cloud-masked fractional cover using load_ard
ds_ls7 = load_ard(dc=dc,
              products=['ls7_sr'],
              like=ds_ls8.geobox,
              time=time,
              measurements=['red','nir'],
              min_gooddata=0.95,
              mask_filters=(['opening',5], ['dilation',5]),
              dask_chunks=dask_chunks
             )
```

```python
cm = dc.load(product=crop_mask,
            time=('2019'),
            measurements='mask',
            resampling='nearest', 
            like=ds_ls8.geobox).mask.squeeze()

cm.plot.imshow(add_colorbar=False, figsize=(5,5))
plt.title('Cropland Extent');
```

### Mask non-croplands

```python
ds_ls8 = ds_ls8.where(cm, np.nan)
ds_ls7 = ds_ls7.where(cm, np.nan)
```

## Match Sentinel-2 and Landsat acquistion times

Tolerance of 7 days


```python
ds_ls7 = ds_ls7.reindex(time=ds_ls8.time, method='nearest', tolerance='3D')
```

```python
ds_ls8 = calculate_indices(ds_ls8, 'NDVI', collection='c2', drop=False)
ds_ls7 = calculate_indices(ds_ls7, 'NDVI', collection='s2', drop=False)
```

### Harmonize Landsat 8 with landsat 7 using Chastain et. al 2019

https://www.sciencedirect.com/science/article/pii/S0034425718305212?via%3Dihub

```python
# multiplying the red band with slope and adding to intercept
slope = 1.0199
intercept = -0.0021
ds_ls7['red_chastain']  = (ds_ls7['red'] * slope) + intercept
```

```python
# multiplying the nir band with slope and adding to intercept
slope = 1.1406
intercept = -0.0163
ds_ls7['nir_chastain']  = (ds_ls7['nir'] * slope) + intercept
```

#### Recalcuate NDVI

```python
ds_ls7['ndvi_chastain'] = (
    (ds_ls7['nir_chastain']-ds_ls7['red_chastain'])
    /
    (ds_ls7['nir_chastain']+ds_ls7['red_chastain'])
)
```

### Reduce to 1D for plotting

```python
ndvi_ls8 = ds_ls8.NDVI.mean(dim=['x','y']).compute()
ndvi_ls7 = ds_ls7.NDVI.mean(dim=['x','y']).compute()
ndvi_ls7_chastain = ds_ls7.ndvi_chastain.mean(dim=['x','y']).compute()
```

### time series plot

```python
plt.figure(figsize=(18, 4))
ndvi_ls8.plot(marker='.', lw=0, color='red', ms=10, label='Landsat 8')
ndvi_ls7.plot(marker='*', lw=0, ms=10, color='green', label='Landsat 7')
plt.legend()
plt.title('Before Band Adjustment')
plt.grid()

plt.figure(figsize=(18, 4))
ndvi_ls8.plot(marker='.', lw=0, color='red', ms=10, label='Landsat 8')
ndvi_ls7_chastain.plot(marker='*', lw=0, ms=10, color='green', label='Landsat 7')
plt.legend()
plt.title('After Band Adjustment')

plt.grid()
plt.show()
```

### Harmonize Landsat 8 with landsat 7 using Aus Coeffiecients
https://www.mdpi.com/2072-4292/6/9/7952/htm

```python
# multiplying the red band with slope and adding to intercept
slope = 1.00446
intercept = 0.00274
ds_ls8['red_aus']  = (ds_ls8['red'] * slope) + intercept
```

```python
# multiplying the nir band with slope and adding to intercept
slope = 0.98906
intercept =0.00004
ds_ls8['nir_aus']  = (ds_ls8['nir'] * slope) + intercept
```

```python
ds_ls8['ndvi_aus'] = (
    (ds_ls8['nir_aus']-ds_ls8['red_aus'])
    /
    (ds_ls8['nir_aus']+ds_ls8['red_aus'])
)
```

```python
ndvi_ls8_aus = ds_ls8.ndvi_aus.mean(dim=['x','y']).compute()
```

```python
plt.figure(figsize=(18, 4))
ndvi_ls8.plot(marker='.', lw=0, color='red', ms=10, label='Landsat 8')
ndvi_ls7.plot(marker='*', lw=0, ms=10, color='green', label='Landsat 7')
plt.legend()
plt.title('Before Band Adjustment')
plt.grid()

plt.figure(figsize=(18, 4))
ndvi_ls8_aus.plot(marker='.', lw=0, color='red', ms=10, label='Landsat 8')
ndvi_ls7.plot(marker='*', lw=0, ms=10, color='green', label='Landsat 7')
plt.legend()
plt.title('After Band Adjustment')

plt.grid()
plt.show()
```

### Harmonize Landsat 8 with landsat 7 using Roy Coeffiecients

https://www.sciencedirect.com/science/article/pii/S0034425715302455

```python
ndvi_ls7_adj = 0.0149 + (1.0035 * ds_ls7.NDVI)
```

```python
ndvi_ls7_adj = ndvi_ls7_adj.mean(dim=['x','y']).compute()
```

```python
plt.figure(figsize=(18, 4))
ndvi_ls8.plot(marker='.', lw=0, color='red', ms=10, label='Landsat 8')
ndvi_ls7.plot(marker='*', lw=0, ms=10, color='green', label='Landsat 7')
plt.legend()
plt.title('Before Band Adjustment')
plt.grid()

plt.figure(figsize=(18, 4))
ndvi_ls8.plot(marker='.', lw=0, color='red', ms=10, label='Landsat 8')
ndvi_ls7_adj.plot(marker='*', lw=0, ms=10, color='green', label='Landsat 7')
plt.legend()
plt.title('After Band Adjustment')

plt.grid()
plt.show()
```
